```python
import pandas as pd
import os
```

```python
data_dir = './data'
```

```python
transactions = pd.read_csv(os.path.join(data_dir, 'transactions_cleaned.csv'))
```

```python
transactions.head()
```

```python
transactions['transaction_date'] = pd.to_datetime(transactions['transaction_date'], format='%Y%m%d')
```

```python
transactions.head()
```

```python
sample = transactions.sample(int(5e6))
```

```python
res = sample.groupby(['msno', pd.Grouper(key="transaction_date", freq="1M")])['payment_plan_days'].agg(['mean', 'median', 'count'])
```

```python
temp = sample.groupby(['msno', pd.Grouper(key='transaction_date', freq='6M')])['is_auto_renew'].sum()
```
