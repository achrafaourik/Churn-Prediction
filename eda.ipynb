### Train Data Early EDA

[(Back to top)](#Table-of-Contents)

```python
def plot_repartition(data, column, title=None, colors=None):
    """Plot the distribution of a categorical column. Creates a bar chart and a pie chart."""
    values = data[column].value_counts()
    nbr_classes = data[column].nunique()
    labels = list(values.keys())   
    ticker = labels
    colors = sns.color_palette('Set3', n_colors=len(ticker))
    cmap = dict(zip(ticker, colors)) 
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
    sns.countplot(x=column, data=data, ax=axes[0], palette=cmap);
    fig.suptitle(title, fontsize=16)
    axes[0].set_ylabel('Total Number')
    axes[1].pie(values, explode=[0.1]*nbr_classes, labels=labels, autopct='%1.1f%%', colors=colors,
            shadow=True, startangle=90)
    axes[1].axis('equal')
    plt.legend(labels, loc='upper right')
    plt.tight_layout()
    plt.show()
```

<a id='churn_distribution'></a>

```python
plot_repartition(train, 'is_churn', title='Distribution of Churn')
```

We can see from the graph that the data is **heavily skewed and *unbalanced***. We can take care of this by  **balancing the classes** using either or combining different of the following methods: **larger weights** for the positive class in the cost function, **under-sampling** the majority class or **over-sampling** the minority class (using **SMOTE** for example). 

If the data is kept as is, the trained models may be biased against the minority class in their predictions and only predict the majority class which is undesirable since in our use case we are interested more in detecting the clients that will churn (meaning the positive class).

```python
def bar_plot_percentage_count(data, column, title=None, xlabel=None, top=None, tick_values=1e6, order=None):
    """Plot a sorted bar plot with twin axes: percentage and count"""    
    temp_data = data.copy()
    top_ids = None
    if xlabel is None:
        xlabel = f'Number of {column}s'
    if top is not None:
        top_ids = temp_data[column].value_counts().index[:top]
    plt.figure(figsize=(12,5))
    ncount = temp_data.shape[0]
    if top_ids is not None:
        ax = sns.countplot(x=column, data=temp_data, order=list(top_ids))
    else:
        if order is None:
            ax = sns.countplot(x=column, data=temp_data, order=temp_data[column].value_counts().index)
        else:
            ax = sns.countplot(x=column, data=temp_data, order=order)
    plt.title(title)
    plt.xlabel(xlabel)

    # Make twin axis
    ax2=ax.twinx()

    # Switch so count axis is on right, frequency on left
    ax2.yaxis.tick_left()
    ax.yaxis.tick_right()
    ax.yaxis.set_label_position('right')
    ax2.yaxis.set_label_position('left')

    ax2.set_ylabel('Frequency [%]')

    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
                ha='center', va='bottom') # set the alignment of the text

    ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_values))

    # Fix the frequency range to 0-100
    ax2.set_ylim(0,100)
    ax.set_ylim(0,ncount)

    ax2.yaxis.set_major_locator(ticker.MultipleLocator(10))
    ax2.grid(None)
    
    plt.show()
```

## Null Values Imputation

### Members Data

[(Back to top)](#Table-of-Contents)

```python
members.isnull().agg(['mean', 'sum']).T.add_prefix('null_')
```

We can see that **more than half** of the '**gender**' column contains null values. Since it might be a good feature of our model, it might be better to fill the values.

Since it is a <ins>categorical</ins> variable, we'll use the **mode** as a *filling value*.

```python
members['gender'].fillna(members['gender'].mode()[0], inplace=True)
```

The same goes for the '**bd**' column (**age**), more than half of the column is empty. It might be useful for the model, so we'll keep it for now.

As for the imputation of missing values in the age column, we will use the median age per gender.

```python
members['bd'].fillna(members.groupby('gender')['bd'].transform('median'), inplace=True)
```

```python
def plot_evolution_categ_monthly(data, date_column, categ_column, format='%Y%m%d'):
    """Plot the monthly evolution of a time series column"""
    data[date_column] = pd.to_datetime(data[date_column], format='%Y%m%d')
    data['date_month'] = data[date_column].dt.year * 100 + data[date_column].dt.month
    temp = data.groupby(['date_month', categ_column])[uid].count().reset_index(name='Count')
    temp['date_month'] = pd.to_datetime(temp['date_month'], format="%Y%m")
    fig1 = alt.Chart(temp).mark_area().encode(
            x=alt.X('yearmonth(date_month):O', title='Month'),
            y=alt.Y("Count:Q", stack="normalize", title='Percentage'),
            color=categ_column + ":N")
    fig2 = alt.Chart(temp).mark_area().encode(
            x=alt.X('yearmonth(date_month):O', title='Month'),
            y=alt.Y("Count:Q", stack=None, title="Count"),
            color=categ_column + ":N")
    return fig1, fig2
```

```python
def generate_data_grouped_plot(data, x_column, grouped_column):
    data_count = data.groupby([x_column, grouped_column])[data.columns[0]].count().reset_index(name="count")
    data_percentage = data.groupby([x_column])[grouped_column].apply(lambda x: x.value_counts(normalize=True)).reset_index(name='percentage').rename(columns={'level_1': grouped_column})
    data_percentage = round(data_percentage, 2)
    data_percentage[grouped_column] = data_percentage[grouped_column].astype(str)
    data_count[grouped_column] = data_count[grouped_column].astype(str)
    return data_count, data_percentage
```

```python
def plot_grouped_bar_chart(data, x_column, y_column, color_column, grouped_column, sort=None, with_text=True, sort2=None):
    bars = alt.Chart().mark_bar().encode(
    x=alt.Column(x_column, sort=sort2),
    y=alt.Column(y_column),
    color=alt.Column(color_column, sort=sort2))

    texts = bars.mark_text(
        align='center',
        baseline='bottom',
        dx=3  
    ).encode(
        text=y_column)
    
    if with_text:
        chart = alt.layer(bars, texts, data=data).facet(
            column=alt.Column(grouped_column, sort=sort,
                              header=alt.Header(labelAngle=0, labelAlign='right'))).properties(title='test')
        return chart.configure_title(fontSize=20, anchor='middle')
    else:
        return alt.layer(bars, data=data).facet(column=alt.Column(grouped_column, sort=sort,
                                                                 header=alt.Header(labelAngle=-90, labelAlign='right')))
```

The distribution of subscriptions is very similar for the whole weekdays and we don't really notice any peak or dip at any given weekday.
